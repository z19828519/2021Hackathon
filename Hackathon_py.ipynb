```python
import torch
import torch.nn as nn
```

```python
from sklearn.model_selection import train_test_split

X = train_X_scaled
y = train_y

X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(X, y, test_size = 0.2)
```

```python
print(X.shape)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        
        self.hidden1 = nn.Linear(52936,300)
        self.hidden2 = nn.Linear(300, 150)
        self.hidden3 = nn.Linear(150, 75)
        self.hidden4 = nn.Linear(75, 30)
        self.output = nn.Linear(30, 1)
        self.dropout = nn.Dropout(0.5)
        self.Tanh = nn.Tanh()
        
    def forward(self,x):
        x = self.hidden1(x)
        x = self.dropout(x)
        x = self.Tanh(x)
        
        x = self.hidden2(x)
        x = self.dropout(x)
        x = self.Tanh(x)
        
        x = self.hidden3(x)
        x = self.dropout(x)
        x = self.Tanh(x)
        
        x = self.hidden4(x)
        x = self.dropout(x)
        x = self.Tanh(x)

        x = self.output(x)
        return x
    
    def predict(self,x):
        y_pred = self.forward(x).detach().numpy()
        #print(y_pred)
        
        return y_pred
    

net = Net()
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr = 0.001)
epochs = 1000
```

```python
#print(X_test)
```

```python
X_train = torch.from_numpy(X_train_raw).type('torch.FloatTensor')
y_train = torch.from_numpy(y_train_raw.to_numpy()).type('torch.FloatTensor')
X_test = torch.from_numpy(X_test_raw).type('torch.FloatTensor')
y_test = torch.from_numpy(y_test_raw.to_numpy()).type('torch.FloatTensor')
```

```python
losses = []
train_MSEs = []
test_MSEs = []
```

```python
import warnings
```

```python
warnings.filterwarnings("ignore")
for i in range(epochs):
    y_pred = net.forward(X_train)
    
    train_pred = net.predict(X_train)
    
    train_MSE = mean_squared_error(y_train, train_pred)
    train_MSEs.append(train_MSE)
    
    test_pred = net.predict(X_test)
    test_MSE = mean_squared_error(y_test, test_pred)
    test_MSEs.append(test_MSE)
    
    loss = criterion(y_pred, y_train)
    print("epoch:", i, "train_MSE:", train_MSE)
    print("test_MSE:", test_MSE)
    print()
    
    losses.append(loss)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
plt.plot(train_MSEs)
plt.plot(test_MSEs)
plt.show()
```
